### Activity 15: Managing Study Groups

You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

**Example data:**

```python
{"John": ["Math", "Physics"], "Emily": ["Physics", "Biology"], "Alex": ["Math"]}

Steps:

1. Choose a data structure that allows quick access to the groups each student belongs to.
2. Write a program that lets students join and leave study groups.
3. Implement a function to display all groups a student is part of.
<hr>

`````python
# Functions to add/remove students from study groups and display their groups
study = {"John": {"Math", "Physics"}, "Emily": {"Physics", "Biology"}, "Alex": {"Math"}}

def join(s, g): study.setdefault(s, set()).add(g)
def leave(s, g): study.get(s, set()).discard(g)
def show(s): print(f"{s}: {', '.join(study.get(s, [])) or 'no groups'}")

# Example usage of functions
join("John", "Chemistry")
leave("Emily", "Biology")
show("John"); show("Emily")

`````

### Activity 17: Counting Words in Les_Brown.txt

Read the content from `Les_Brown.txt`, and then count the number of occurrences of each word in the text. Write the word count results to a new file called `word_count.txt`.

**Steps:**
1. Open and read the contents of `Les_Brown.txt`.
2. Split the content into words and count the frequency of each word.
3. Write the word count in the format: `word: count` to a new file called `word_count.txt`.

<hr>

`````python
# Read the text file, clean punctuation, convert to lowercase, and count word occurrences
from collections import Counter
import string

with open("Les_Brown.txt") as f:
    words = f.read().lower().translate(str.maketrans('', '', string.punctuation)).split()

counts = Counter(words)

# Write the word counts to a new file and print them
with open("word_count.txt", "w") as f:
    f.writelines([print(f"{w}: {c}") or f"{w}: {c}\n" for w, c in counts.items()])

`````

### Activity 19 (Set): Splitting Clothing Stock by Size

Create a Python program that reads the `clothing_stock.txt` file and splits the clothing items into separate sets based on their size (e.g., all XS items in one set, all M items in another set). Print each set.

**Steps:**
1. Read the file and separate items by their size.
2. Store each size in a different set.
3. Print each set of items.
<hr>

`````python
#Suriya kiran. S
#MISIS no-M01046518
from collections import defaultdict  # Import defaultdict to automatically create sets for new keys

size_dict = defaultdict(set)  # Create a dictionary where each key (size) maps to a set of items

with open("clothing_stock.txt", "r") as file:  # Open the clothing stock file in read mode
    for line in file:
        item, size = line.strip().split(" ")  # Split each line into item name and size
        size_dict[size].add(item)  # Add the item to the corresponding size set

for size in sorted(size_dict.keys()):  # Sort sizes alphabetically and print the items for each
    print(f"{size}: {size_dict[size]}")
`````
